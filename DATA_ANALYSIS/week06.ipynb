## 1. 평균 거래액? 
월별 평균 거래액 분석

```python
orders.head()
```

```python
order_items.head()
```

```python
payments.head()
```

```python
#없는 데이터 확인
orders.isnull().sum()
```

```python
# 없는 데이터 삭제 
orders=orders.dropna()
```

```python
orders.info()
```

```python
# payments는 없는 데이터가 없는 상태 
payments.isnull().sum()
```

```python
# orders와 payments 사이즈 비교 
payments.info()
```

```python
orders.info()
```

```python
# order_id 중복된 데이터가 있는지 확인 
orders['order_id'].value_counts().max() #없음 
```

```python
payments['order_id'].value_counts().max() #있음 
```

```python
payments['order_id'].value_counts()
```

```python
payments[payments['order_id']=='fa65dad1b0e818e3ccc5cb0e39231352']
```

```python
payments=payments.groupby('order_id').sum()
payments[payments.index=='fa65dad1b0e818e3ccc5cb0e39231352']
```

```python
# orders의 구매 날짜와 payments의 총 지불 금액을 합침 
merged_order=pd.merge(orders,payments,on='order_id',how='left')
```

```python
merged_order.info()
```

```python
merged_order[merged_order['order_id']=='fa65dad1b0e818e3ccc5cb0e39231352']
```

### pandas로 날짜 다루기 
시계열 자료 이용해야하므로(order_purchase_timestamp,..)  

```python
merged_order.info()
```

```python
merged_order['order_purchase_timestamp']=pd.to_datetime(merged_order['order_purchase_timestamp'],format='%Y-%m-%d %H:%M:%S',errors='raise')
```

```python
merged_order.info()
```

```python
#데이터 프레임 중 일부를 선택 후, 조작하면 해당 데이터프레임도 변경 
#copy를 통해 복사본을 만들어 조작하여, 원본 데이터프레임은 보존 가능 
merged_order_payment_date=merged_order[['order_purchase_timestamp','payment_value']].copy()
```

```python
merged_order_payment_date.head()
```

```python
merged_order_payment_date=merged_order_payment_date.set_index('order_purchase_timestamp')
```
